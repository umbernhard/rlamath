So far we've covered some important ground. We know we can frame election outcomes as binomial processes, and we can capture invalid votes. With this knowledge, we're ready to start asking questions about election outcomes. 

Let's imagine an election with 1,000 ballots in it, where Alice gets 600 votes, Bob gets 300 votes, and 100 votes are invalid. Additionally, we draw a sample of 50 ballots: 30 are for Alice, 15 are for Bob, and 5 are invalid. Given this information, how likely is it that Alice really won our election?

To answer this question, we're going to need to know a few things about probability. Before, when we were examining the permutations of all possible elections, we asked how likely it was that Alice lost given some sample data. This is a **conditional probability**. If $A$ is all elections where Alice won, $A^c$ is all of the elections where she didn't win. $A^c$ is the **complement** of $A$, such that an election $\vec{e}$ in $A$ is not in $A^c$ and vice-versa. An important thing to note is that the probability an election is in $A$ is the same as the probability that it isn't in $A^c$, i.e. $P(\vec{e} \in A) = 1 - P(\vec{e} \in A^c$).

If we don't have any information about the election, we saw that Alice was just as likely to win as Bob, so $P(\vec{e} \in A) = P(\vec{e} \in A^c)$. For notational simplicity, I'm going to write $P(\vec{e} \in A)$ simply as $P(A)$, and the same for $A^c$. Just remember that when I say $P(A)$, what I really mean is the probability that an election is in $A$. 

In our 10 ballot election before, we saw that after 6 ballots, if 5 were for Alice and 1 was for Bob, the probability that Alice won was around 93%. We can treat our 6 ballots as a sample $S$. The probability we calculated before is a conditional probability $P(A|S)$ (read as "the probability an election is in $A$ given a sample $S$"). The distribution we showed with the given sample was a conditional distribution.
